```python
import pandas as pd

df = pd.read_csv('엔솔_전처리_파단직전재추출_0401.csv', encoding='utf-8')

df
```

```python
## save seperate dataframs base on the 'error_index'


for i in df['error_index'].unique():
    df_temp = df[df['error_index'] == i]
    df_temp.to_csv(f'엔솔_전처리_파단직전재추출_0401_{i}.csv', index=False, encoding='utf-8')
    print(f'엔솔_전처리_파단직전재추출_0401_{i}.csv')
    print(df_temp)
```

```python
df['ds'] = pd.to_datetime(df['ds'], format='%Y-%m-%d %H:%M:%S')
```

```python
import warnings
warnings.filterwarnings('ignore')

for i in df['error_index'].unique():
    df_temp = df[df['error_index'] == i]
    
    # Check if there are any fault conditions for this error_index
    fault_rows = df_temp[df_temp['파단직전'] == 1]
    
    if len(fault_rows) > 0:
        # Find the timestamp where 파단직전 = 1
        fault_time = fault_rows['ds'].iloc[0]
        
        # Find the first timestamp in this segment
        start_time = df_temp['ds'].min()
        
        # Calculate time difference until fault (from start to fault)
        time_to_fault = (fault_time - start_time).total_seconds()
        
        print(f'Error Index: {i}')
        print(f'Total seconds until fault: {time_to_fault} seconds')
        print('-' * 50)
    else:
        print(f'Error Index: {i}')
        print('No fault condition found in this segment')
        print('-' * 50)
```
